imprting libraries

```python
import numpy as np
import pandas as pd
import plotly.figure_factory as ff
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

reading data

```python
df = pd.read_csv("coins_data.csv")
```

```python
df
```

```python
df.info()
```

```python
x = df['Volume'] 
y = df['MarketCap']
```

```python
data = list(zip(x, y))
```

clustering dendrogram plot for market cap and volume

```python
linkage_data = linkage(data, method='ward', metric='euclidean')
plt.figure(figsize=(20,7))
dendrogram(linkage_data, labels=df["Symbol"].tolist(), leaf_font_size=10)
plt.xlabel("Cryptocurrency")
plt.ylabel("Distance")
plt.show()
```

clustering algorithm

```python
from sklearn.cluster import AgglomerativeClustering

hierarchical_cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
labels = hierarchical_cluster.fit_predict(data)

plt.figure(figsize=(20,7))
plt.scatter(x, y, c=labels)
for i, txt in enumerate(df['Symbol']):
    plt.annotate(txt, (x[i], y[i]))
plt.show()
```

```python
for i in ['Volume', 'MarketCap']:
    plt.figure(figsize=(20, 7))
    plt.bar(df['Symbol'], df[i])
    plt.xlabel('Symbol')
    plt.ylabel(i)
    plt.show()
```

clustring with market cap volume and proof type

```python
label_encoder = LabelEncoder()
z = label_encoder.fit_transform(df['ProofType'])
```

```python
data = list(zip(x, y, z))
```

```python
linkage_data = linkage(data, method='ward', metric='euclidean')
plt.figure(figsize=(20,7))
dendrogram(linkage_data, labels=df["Symbol"].tolist(), leaf_font_size=10)
plt.xlabel("Cryptocurrency")
plt.ylabel("Distance")
plt.show()
```

```python
hierarchical_cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
labels = hierarchical_cluster.fit_predict(data)

plt.figure(figsize=(20,7))
plt.scatter(x, y, c=labels)
for i, txt in enumerate(df['Symbol']):
    plt.annotate(txt, (x[i], y[i]))
plt.show()
```
